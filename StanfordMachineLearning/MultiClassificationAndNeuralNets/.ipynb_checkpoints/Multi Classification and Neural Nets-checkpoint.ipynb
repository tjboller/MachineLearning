## MultiClass Logistic Classification

$\\$

The data is in a matlab file, luckily scipy has a function for that (https://xkcd.com/353/)

```python
data = sio.loadmat('ex3data1.mat')
```

$\\$ 

The data has the training images in a matrix called 'X'. It is a 5000x400 matrix where each row is a raveled 20x20 image of a hand drawn number. The data also comes with a 5000x1 vector with the actual digit.

```python
trainingImages = data['X']

# Send every element with 10 to 0 -> original data uses ten because of matlab indexing
# digit = np.array([0 if elem == 10 else elem for elem in np.concatenate(data['y'])])
digit = np.array([[0] if elem == 10 else elem for elem in data['y']])
```

$\\$

First we will classify all the handwritten digits using the 'one versus rest' logistic regression technique. Scikit-learn actually has this built into their logistic regression class. However, the whole point of going through these exercises is to learn! I already implemented my own binary logistic qualifier in another project, so I used scikit-learns binary logistic regression to implement my own 'one versus rest' qualifier.

```python
def ova(X,y):
    
    # Initialize model
    model = lm.LogisticRegression()
    
    # Initialize max values
    records = np.zeros(len(y))
    fit = np.zeros(len(y))
    
    # For each unique label (class) in the observations
    for label in np.unique(y):
        
        # Fit a regression for the class, and predict the probabilities (only grabbing prob that x is pos)
        new_fit = model.fit(X, [1 if obsv == label else 0 for obsv in y]).predict_proba(X)[:,1]       
        
        for i in range(len(y)):
            
            # Replace records/fit with new records
            if new_fit[i]>records[i]:
                records[i] = new_fit[i]
                fit[i] = label
                
    return fit
```


$\\$

Lets run and time the regressions and see how my implementation compares! It's not too much slower and the fit is the same, success!

```python
%timeit ova(trainingImages, digit.ravel())
custom = ova(trainingImages, digit.ravel())
```

```python
model = lm.LogisticRegression()
%timeit model.fit(trainingImages, digit.ravel()).predict(trainingImages)
builtin = model.fit(trainingImages, digit.ravel()).predict(trainingImages)
```

```python
sum(builtin != custom)
```

$\\$

Overall the method worked pretty well, a 94% success rate

```python
model.score(trainingImages, digit.ravel())
```

$\\$

If possible it is always good pratice to explore your result. We plot a sample of the numbers we did not classify correctly using logistic regression.  IDK about you, but I could definitely see how some of these were misclassified. 

```python
# Get the rows that were classified incorrectly (logical indexing)
wrong = trainingImages[builtin != digit.ravel(),:]

fig, ax = plt.subplots(5,10, figsize=(16,9))


for i in range(5):
    for j in range(10):
        ax[i,j].grid(False)
        ax[i,j].imshow(wrong[np.random.randint(wrong.shape[0]),:].reshape(20,20).T)
```

## Neural Networks

It is very expensive to train high-order polynomail features of a logistic regression, so the section above uses only linear features and does quite well - a 94% success rate. Neural networks however can learn complex nonlinear hypothesis. The weights of the neural networks have already been calculated (Calculating own weights on next assignment) 

```python
weights = sio.loadmat('ex3weights.mat')
Theta1 = weights['Theta1']
Theta2 = weights['Theta2']
```


$\\$

Calculate the prediction by using forward propagation

```python
sigmoid = sp.expit

# Add a bias term for each example
c_trainingImages = np.insert(trainingImages,0,1,axis=1)

# Initialize list
neural_net = []

# For each image, calculate the classification
for image in c_trainingImages:
    
    # First multiple theta1 and the observation (image), then take sigmoid, then add bias unit, then repeat for second layer
    # append the index of the largest element to the classification array
    neural_net.append(np.argmax(sigmoid(Theta2.dot(np.insert(sigmoid(Theta1.dot(image)),0,1)))))

# Convert matlabs labeling to our own
neural_net = np.array([(label+1)%10 for label in neural_net])
```


$\\$

Calculate the percent of images classified correctly, not surprisingly the neural network performs better than the linear logistic regression. - 97%

```python
1 - sum(neural_net!=digit.ravel())/len(neural_net)
```


$\\$

Again, let's check out the ones it didn't get right. By inspection, these do seem harder to classify than the sample from the logistic regression

```python
# Get the rows that were classified incorrectly (logical indexing)
wrong = trainingImages[neural_net != digit.ravel(),:]

fig, ax = plt.subplots(5,10, figsize=(16,9))


for i in range(5):
    for j in range(10):
        ax[i,j].grid(False)
        ax[i,j].imshow(wrong[np.random.randint(wrong.shape[0]),:].reshape(20,20).T)
```
